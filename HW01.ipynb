```python
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-05/villagers.csv"
df = pd.read_csv(url)

# Display column names and their data types
print(df.dtypes)

# Summary statistics for numeric columns
print(df.describe())

# Summary of unique values and their counts for categorical columns
for column in df.select_dtypes(include=['object']).columns:
    print(f"\nColumn: {column}")
    print(df[column].value_counts())

# Check for missing values
print(df.isnull().sum())

```

```python
# df.describe() only displays the numeric columns so there is a discrepancy from df. shape which reports total number of columns, including categarical as well
```

```python
# attribute is any piece of information describing an object, doesn't use () since it doesnt carry out any process
# method is a function in programming which carries out a process to fetch different information sets and display them or conduct calculations to summarize the data
```

```python
"""
1. Understanding df.shape vs. df.describe()
df.shape (Attribute)

Definition: An attribute that provides the dimensions of a DataFrame. It returns a tuple (number_of_rows, number_of_columns).
Usage: Accessed directly without parentheses. Example: df.shape returns (100, 10) indicating 100 rows and 10 columns.
df.describe() (Method)

Definition: A method that provides summary statistics for numeric columns in a DataFrame. It returns statistics such as mean, standard deviation, min, max, and percentiles.
Usage: Called with parentheses. Example: df.describe() returns a table with statistical summaries for numeric columns.
Key Difference:

Attribute: Provides direct information about the DataFrame's structure.
Method: Performs computations or actions and returns results based on the data.
2. Finding the Number of Rows and Columns in a DataFrame
Method Used: df.shape
Explanation: df.shape returns the dimensions of the DataFrame. For example, df.shape might return (100, 10), indicating 100 rows and 10 columns.
3. Providing a Simple Summary of Columns
Numeric Columns:

Method Used: df.describe()
Explanation: Provides statistical summaries for numeric columns, including mean, standard deviation, min, max, and quartiles.
Categorical Columns:

Method Used: df.describe(include='object')
Explanation: Provides summaries for categorical columns, including counts of unique values, the most frequent value, and its frequency.
Missing Values:

Method Used: df.isnull().sum()
Explanation: Identifies missing values in each column.
4. Difference Between Mathematical Functions and Programming Functions
Mathematical Function:

Definition: A relation that assigns exactly one output to each input, such as 
f(x)=x^2
 .
Nature: Abstract and theoretical.
Programming Function/Method:

Definition: A piece of code that performs a specific task or calculation when called, such as df.describe().
Nature: Concrete and executable; performs operations on data.
Example:

Mathematical Function: 
𝑓(𝑥)=𝑥2, where 𝑥=3 results in 9.
Programming Method: df.describe() computes and returns statistical summaries for numeric columns in a DataFrame.
"""
```

```python
# count- The number of values in a column that are not missing or null.
# mean- The average value of all the numerical data in the column, a measure of central tendency.
# std(standard deviation)- Square root of variance, it is a measure of the spread of numerical data in the column.
# min(minimum)- Smallest numerical value in the column, helps to measure range of data.
# 25%(25th percentile)- Lower quartile, the data at the 25% percent mark of the population.
# 50%(50th percentile)- Median, the data at the middle of the population when in ascending order.
# 75%(75th percentile)- Upper quartile, the data at the 75% percent mark of the population.
# max(maximum)- Highest numerical value of the column, helps to measure the range.
```

```python
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-05/villagers.csv'
df = pd.read_csv(url)

# Remove unnecessary columns
if 'song' in df.columns:
    df_cleaned = del df['song']

# Remove rows with missing data
df_cleaned = df_cleaned.dropna(axis=0, how='any')

print(df_cleaned)
```

```python
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-05/villagers.csv'
df = pd.read_csv(url)

# Remove unnecessary columns
if 'song' in df.columns:
    del df['song']

# Remove rows with missing data
df_cleaned = df.dropna(axis=0, how='any')

print(df_cleaned)
```
